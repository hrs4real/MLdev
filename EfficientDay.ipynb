```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv("Task.csv")
data
```

```python
new_data= (data.drop('Name', axis=1)).drop('Breakfast', axis=1)
new_data
```

```python
data.columns
```

```python
Breakfast=data['Breakfast']
dummy_state=pd.get_dummies(Breakfast,drop_first=True)
dummy_state
```

```python
new_data['Yes']=dummy_state
new_data
```

```python
model=LinearRegression()
```

```python
y=new_data['How efficient you are in next day [ Scale 1 - 100 ]']
```

```python
x=new_data.drop('How efficient you are in next day [ Scale 1 - 100 ]',axis=1)
X_train,X_test,Y_train,Y_test= train_test_split(x,y,test_size=0.20,train_size=0.80,random_state=42)
```

```python
X_train
```

```python
x
```

```python
model.fit(X_train,Y_train)
```

```python
model.coef_
```

```python
Y_pred=model.predict(X_test)
Y_pred
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(Y_test,Y_pred)
```
